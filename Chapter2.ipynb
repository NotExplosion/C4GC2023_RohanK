```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data and converting it to a date-time format
fin_data = pd.read_csv("tute1.csv")
fin_data['Quarter'] = pd.to_datetime(fin_data['Quarter'], format='%Y-%m')
fin_data.index = fin_data['Quarter']
del fin_data['Quarter']
fin_data.head()
```

```python
#line plot of the three time series in the data set
sns.lineplot(fin_data)
plt.ylabel("Dollars")
```

```python
#line plot of the three time series individually
fig, axes = plt.subplots(3,1)
sns.lineplot(data=fin_data, x='Quarter', y='Sales', ax=axes[0])
sns.lineplot(data=fin_data, x='Quarter', y='AdBudget', ax=axes[1])
sns.lineplot(data=fin_data, x='Quarter', y='GDP', ax=axes[2])
```

```python
#autocorrelation plot of sales
pd.plotting.autocorrelation_plot(fin_data[['Sales']])
```

```python
#autocorrelation plot of advertising budget
pd.plotting.autocorrelation_plot(fin_data[['AdBudget']])
```

```python
#autocorrelation plot of GDP
pd.plotting.autocorrelation_plot(fin_data[['GDP']])
```

```python
#Seasonal additive decomposition, for sales
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(fin_data['Sales'], model = 'additive')
result.plot()
```

```python
#Seasonal multiplicative decomposition, for GDP
result2 = seasonal_decompose(fin_data['GDP'], model = 'multiplicable')
result2.plot()
```

```python
#Seasonal multiplicative decomposition, for advertising budget
result3 = seasonal_decompose(fin_data['AdBudget'], model = 'multiplicable', period=4)
result3.plot()
```
