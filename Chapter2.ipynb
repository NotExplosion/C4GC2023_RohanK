```python
#File Download Link: https://bit.ly/fpptute1
#Necessary imports for this chapter:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing data
fin_data = pd.read_csv("tute1.csv")

#Replaces the default index with a time series index
fin_data['Quarter'] = pd.to_datetime(fin_data['Quarter'], format='%Y-%m')
fin_data.index = fin_data['Quarter']
del fin_data['Quarter']

fin_data.head()
```

```python
#Line plot of the three time series in the data set
sns.lineplot(fin_data)
plt.ylabel("Dollars")
```

```python
#Line plot of the three time series individually
fig, axes = plt.subplots(3,1)
sns.lineplot(data=fin_data, x='Quarter', y='Sales', ax=axes[0])
sns.lineplot(data=fin_data, x='Quarter', y='AdBudget', ax=axes[1])
sns.lineplot(data=fin_data, x='Quarter', y='GDP', ax=axes[2])
```

```python
fig, axes2 = plt.subplots(3,1)
#autocorrelation plot of sales
pd.plotting.autocorrelation_plot(fin_data[['Sales']], ax=axes2[0])
#autocorrelation plot of advertising budget
pd.plotting.autocorrelation_plot(fin_data[['AdBudget']], ax=axes2[1])
#autocorrelation plot of GDP
pd.plotting.autocorrelation_plot(fin_data[['GDP']], ax=axes2[2])
```

```python
#File download link: https://bit.ly/fpptourism
#Reading in a data set with multiple time series
tourism = pd.read_excel("tourism.xls", index_col=None, header=0)

#Changing the index format to date-time
tourism['Quarter'] = pd.to_datetime(tourism['Quarter'], format='%Y-%m')
tourism.index = tourism['Quarter']
del tourism['Quarter']
tourism.head()
```

```python
#Grouping the data points by region
regional_tourism = tourism.groupby("Region")

#Finding the purpose with the greatest number of trips, based on region
max_tourism = regional_tourism.max()


#Finding the combination of region and purpose with the greatest number of trips overall
abs_max = max_tourism.max()
print(abs_max)

max_tourism.head()
```
