```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#function to calculate the total mean and standard deviation of an entire time series
def ts_meanstd(ts):
  mean = round(ts.mean(axis=0), 3)
  std = round(ts.std(axis=0), 3)
  print("Mean:", mean, ", Standard Deviation:", std)
  return mean, std
```

```python
#importing data; making a copy of the unmodified data for later use

fin_data = pd.read_csv("tute1.csv")
fin_data['Quarter'] = pd.to_datetime(fin_data['Quarter'], format='%Y-%m')
fin_data2 = fin_data.copy()
fin_data2.head()

```

```python
#replacing the standard index of the dataframe with a date-time index

fin_data.index = fin_data['Quarter']
del fin_data['Quarter']
fin_data.head()
```

```python
#finding the total mean and standard deviation of the data, as well as the rolling mean with a window of size 4

print("Sales Data:")
s_mean, s_std = ts_meanstd(fin_data['Sales'])
print("\nAdvertising Budget Data:")
a_mean, a_std = ts_meanstd(fin_data['AdBudget'])
print("\nGDP Data:")
g_mean, g_std = ts_meanstd(fin_data['GDP'])
fin_data.rolling(4).mean()

```

```python
#rolling standard deviation of the data with a window of size 4
fin_data.rolling(4).std()
```

```python
#reading in a data set with multiple time series#
tourism = pd.read_excel("tourism.xls", index_col=None, header=0)

#changing the index format to date-time
tourism['Quarter'] = pd.to_datetime(tourism['Quarter'], format='%Y-%m')
tourism.index = tourism['Quarter']
del tourism['Quarter']
tourism.head()
```

```python
#grouping the data points by region
regional_tourism = tourism.groupby("Region")

#finding the purpose with the greatest number of trips, based on region
max_tourism = regional_tourism.max()

#finding the combination of region and purpose with the greatest number of trips overall
abs_max = max_tourism.max()
print(abs_max)
max_tourism.head()
```

```python
#Forecasting implementation
from statsforecast import StatsForecast
from statsforecast.models import Naive, SeasonalNaive, RandomWalkWithDrift

#creating the forecasting model, using simple methods (Naive and Seasonal Naive, was unable to implement Random Walk with Drift)
model = StatsForecast(models=[Naive(), SeasonalNaive(season_length=1)], freq='QS', n_jobs=-1)
#print(fin_data[['Sales']])

#for the model to work, the data has to be in a specific format
#'unique_id' is a value that represents which time series a data point belongs to
#in this case, all data is part of the same time series so should have the same 'unique_id'
#'ds' is the date/time column; 'y' is the value being forecasted
fin_data2 = fin_data2.rename({'Quarter':'ds','Sales':'y'}, axis=1)
fin_data2['unique_id'] = 1
fin_data2.head()

#h is the number of data points into the future that the model is forecasting
model.forecast(h=8, df=fin_data2)
```
